```python
from torch import torch


def tensors_on_gpus():
    x = torch.zeros(32, 32)
    print(x)
    print(x.device)
    print(torch.device("mps"))
    print(torch.cuda.is_available())
    print(torch.cuda.device_count())
    print(torch.cuda.get_device_properties(0))


tensors_on_gpus()
```

```python
def tensor_storage():
    x = torch.zeros(4, 3)
    print(x)
    r, c = 1, 2
    index = r * x.stride(0) + c * x.stride(1)
    print(index)


tensor_storage()
```

```python
def tensor_slicing():
    x = torch.tensor([[1, 2, 3], [4, 5, 6]])
    print(x[0])
    print(x[:, 0])
    print(x.view(3, 2))
    print(x.transpose(1, 0))


tensor_slicing()
```

```python
B = 1024
D = 256
K = 64
x = torch.ones(B, D)
w = torch.randn(D, K)
y = x @ w
y
actual_num_flops = 2 * B * D * K
```

```python
def gradient():
    x = torch.tensor([1.0, 2, 3])
    w = torch.tensor([1.0, 1, 1], requires_grad=True)
    pred_y = x @ w

    loss = 0.5 * (pred_y - 5).pow(2)
    print(f"loss before backward: {loss}")
    print(f"loss.grad before backward: {loss.grad}")
    print(f"x.grad before backward: {x.grad}")
    print(f"pred_y.grad before backward: {pred_y.grad}")
    print(f"w.grad before backward: {w.grad}")
    loss.backward()
    print(f"loss after backward: {loss}")
    print(f"loss.grad after backward: {loss.grad}")
    print(f"x.grad after backward: {x.grad}")
    print(f"pred_y.grad after backward: {pred_y.grad}")
    print(f"w.grad after backward: {w.grad}")


gradient()
```
